```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.naive_bayes import BernoulliNB
%matplotlib inline
```

```python
df=pd.read_json('Office_Products_5.json',lines=True)
```

```python
df.head()
```

```python
from collections import Counter
word_array=Counter(" ".join(df['summary']).split(" ")).items()
```

```python
word_df=pd.DataFrame(list(word_array))
```

```python
word_df.columns=['word','count']
```

```python
#in order to remove duplicates
word_df['word']=word_df['word'].str.lower()
word_df['word']=word_df['word'].str.strip()
word_df['word']=word_df['word'].str.replace('.', '', regex=False)
word_df['word']=word_df['word'].str.replace('!', '', regex=False)
```

```python
word_df.sort_values('count', ascending=False).head(50)
# great, good, nice, excellent, easy, perfect, love
```

```python
keywords=['great', 'good', 'nice', 'excellent', 'easy', 'perfect', 'love','product','easy','price']
```

```python
features=pd.DataFrame()
```

```python
y=df['overall']
for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    features[str(key)] = df.summary.str.contains(
        (str(key)),
        case=False
    )
data=features[keywords]
```

```python
bnb = BernoulliNB()

bnb.fit(data, y)

y_pred = bnb.predict(data)

print("Number of mislabeled points out of a total {} points : {}".format(
    data.shape[0],
    (y != y_pred).sum()
))
```
